### Calculate Class Average and Status
This code block initializes a list of student marks. It then attempts to calculate the total sum of marks and the average. **Note:** The loop `for i in range(len(marks) - 1)` will iterate one element less than the actual length, causing the last mark to be excluded from the total sum. This will result in an incorrect average. After calculation, it determines if the class passed or failed based on whether the average is above 50.

```python
marks = [45, 60, 72, 38, 90]

total = 0

for i in range(len(marks) - 1):
    total += marks[i]

average = total / len(marks)

if average > 50:
    print("Class passed")
else:
    print("Class failed")

print("Average:", average)
```

### Import Libraries

### Prepare Student Data
This code block defines a dictionary named `data` that contains information for several students, including their ID, attendance percentage, assignment score, quiz score, and whether they completed the lab. This data will be used to create a DataFrame.

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percent": [92, 67, 81, 45, 74, 88],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}
```

### Create DataFrame
This cell converts the `data` dictionary into a pandas DataFrame named `df`. This structured format makes it easier to manipulate and analyze the student data.

```python
df = pd.DataFrame(data)
df
```

### Calculate Total Score
This code block creates a new column named `total_score` in the DataFrame `df`. This score is calculated by summing the `assignment_score` and `quiz_score` for each student.

```python
df['total_score'] = df['assignment_score'] + df['quiz_score']
df
```

### Determine Student Eligibility
This cell adds a new boolean column `eligible` to the DataFrame `df`. A student is considered eligible if their `attendance_percent` is 75% or higher, their `total_score` is 70 or higher, AND they have `lab_completed`.

```python
df['eligible'] = (
    (df['attendance_percent'] >= 75) &
    (df['total_score'] >= 70) &
    (df['lab_completed'])
)
df
```

### Filter for Eligible Students
This code block creates a new DataFrame called `eligible_students` by filtering the original `df` DataFrame. It selects only those rows where the `eligible` column is `True`, effectively listing all students who meet the eligibility criteria.

```python
eligible_students = df[df["eligible"] == True]
eligible_students
```
