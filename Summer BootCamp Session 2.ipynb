This cell imports the necessary libraries: `pandas` for data manipulation and `matplotlib.pyplot` for plotting.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

This cell defines a dictionary `data` containing student information, including `student_id`, `attendance_percent`, `assignment_score`, `quiz_score`, and `lab_completed`.

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percent": [92, 67, 81, 45, 74, 88],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}
```

This cell converts the `data` dictionary into a pandas DataFrame `df`. It then calculates the `total_score` by summing `assignment_score` and `quiz_score`. Finally, it determines student `eligible` status based on attendance, total score, and lab completion.

```python
df = pd.DataFrame(data)


df['total_score'] = df['assignment_score'] + df['quiz_score']

df['eligible'] = (
    (df['attendance_percent'] >= 75) &
    (df['total_score'] >= 70) &
    (df['lab_completed'])
)
```

This cell generates a bar chart showing the `attendance_percent` for each student. The x-axis represents `student_id` and the y-axis represents `Attendance (%)`.

```python
plt.figure(figsize=(6,4))
plt.bar(df['student_id'], df['attendance_percent'])
plt.title("Attendance Percentage")
plt.xlabel("Student ID")
plt.ylabel("Attendance (%)")
plt.show()
```

This cell creates a bar chart visualizing the `total_score` for each student. The x-axis is `Student ID` and the y-axis is `Total Score`.

```python
plt.figure(figsize=(6,4))
plt.bar(df['student_id'], df['total_score'])
plt.title("Total Score of Students")
plt.xlabel("Student ID")
plt.ylabel("Total Score")
plt.show()
```

This cell generates a scatter plot to show the relationship between `assignment_score` and `quiz_score`. Each point represents a student, and their `student_id` is annotated next to their respective point.

```python
plt.figure(figsize=(6,4))
plt.scatter(df['assignment_score'], df['quiz_score'])
plt.title("Assignment vs Quiz Score")
plt.xlabel("Assignment Score")
plt.ylabel("Quiz Score")
```

This cell generates a pie chart showing the distribution of eligible and non-eligible students based on the calculated `eligible` column.

```python
eligible_count = df['eligible'].value_counts()

plt.figure(figsize=(5,4))
plt.pie(eligible_count,
        labels=['Eligible', 'Not Eligible'],
        autopct='%1.1f%%')
plt.title("Eligibility Distribution")
plt.show()
```
